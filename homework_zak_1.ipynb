#### 1. Как бы Вы организовали аналитику по данному продукту?

Возьму на себя смелость слегка переформулировать задание. Видимо, речь все-таки идет о **процессе** получения профессии Data Scientist, а не просто о профессии. Поскольку это - процесс, результат которого равно зависит как от усилий клиента, так и от работы команды Нетологии, задачей аналитики этого процесса является оптимизация процесса формирования специалиста по данной специальности. Ситуация будет тем более идеальной, чем более высококвалифицироваными будут выпускники курса и чем меньше ресурсов для достижения этого результата понадобится вложить Нетологии. Аналитика должна помочь в поиске соответствующих образовательных технологий.
Аналитическая работа должна выполняться постоянно, результаты анализа очередного учебного потока могут приводить к корректировке учебного плана для последующих потоков.

#### 3. Какие данные Вам могут помочь в поиске инсайтов?

1. На этапе привлечения клиентов - изучение конкурентов. 
- Чем привлекательны для клиентов другие платные курсы?

2. На этапе обучения: 
- Лекции каких преподавателей пользуются наибольшей популярностью и почему (возможная ротация преподавателей)?

- Каково качество выполнения домашних работ? 
- Насколько часто студенты обращаются за помощью менеджеров или преподавателей?
- Насколько быстро студенты получают ответы на свои вопросы?
- Отзывы студентов по обучению по каждой из учебных активностей.
- Поиск возможностей объединения студентов разных потоков на общих модулях
3. На этапе финализации обучения:
- Интеграция с бизнесом для выбора тем дипломных проектов;
- Уровень дипломных проектов;
- Качество защиты
4. После обучения:
- Трудоустройство или карьерный рост (в том числе - в Нетологии);
- Продолжение обучения на новых программах Нетологии;
- Новые клиенты, пришедшие по рекомендации закончивших обучение.

#### 4. Опишите команду для работы с данными для данного продукта
- **BI-аналитик** - сбор и обобщение данных; 
- **Аналитик данных** - выбор метрик оценки, эксперименты по изменению рекламных стратегий (для продвижения) и элементов учебного плана (для оптимизации процесса обучения)
- **Маркетолог** - продвижение проекта, поиск клиентов, интервью с клиентами (как один из источников информации)
- **Куратор специальности** - общая поддержка студентов специальности всех потоков, онлайн-общение со студентами, организация внеурочных активностей (митапы и т.п.)
- **Кураторы модулей** - поддержка процесса обучения на конкретном модуле,анализ обратной связи от студентов и преподавателей
- **Преподаватели** - Запись лекций для самостоятельного изучения, подготовка презентаций, проведение занятий и коррекция этого контента на основании результативности обучения, а также появления новых технологий в преподаваемой предметной области
- **Дата-сайентисты** - Прогнозирование выгорания студентов, прогнозирование карьерных результатов, рекомендательная система студентов для работодателей, подбор лучшей цены курса и т.п.
- **Разработчики** - Единая программная платформа с фронтэндом для студентов и бэкендом для организаторов, возможно, с реализацией специфических решений для конкретной профессии.
- **CDO** - (Chief Data Officer) - интеграция данных по специальности с единой информационной системой Нетологии
- **CEO** - Начальство всегда нужно.

#### 6. Чем они будут заниматься?

Конкретное распределение обязанностей рамочно обозначено выше, в п.4.  
1. На первом этапе запуска новой профессии куратор специальности, маркетолог, аналитик и дата-сайентист должны начать проектирование учебного плана с изучения рынка. Одновременно можно изучать рынок учебных курсов и рынок вакансий по данной профессии.
2. После формирования программы курса должны быть подобраны преподаватели, должна быть подсчитана экономика программы.
3. Должны быть определены метрики для продвижения курса.
4. Должна быть установлена стоимость обучения и система скидок, после этого начинается рекламная кампания.
5. Должны быть определены метрики для курса обучения 
6. После начала продаж и формирования первого потока должна быть организована система получения обратной связи по обучению.
7. При выходе метрик за установленные границы должны предприниматься корректирующие воздействия на учебный процесс. Может корректироваться программа, студентам могут предлагаться варианты с переносом сроков обучения и т.п., может производиться смена преподавателей.
8. По мере прохождения обучения и после завершения курса наиболее успешным студентам и выпускникам могут предлагаться варианты участия в проектах Нетологии :)
